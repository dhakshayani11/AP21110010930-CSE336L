```python
import matplotlib.pyplot as plt
```

```python
#Q4

with open('axis_values.txt', 'r') as file:
    data = file.readlines()
    x = [float(val) for val in data[1].split()]
    y = [float(val) for val in data[0].split()]

# Plotting
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')
plt.show()
```

```python
#Q5

x = [1, 2, 3, 4, 5]
y1 = [23, 24, 30, 44, 55]
y2 = [5, 4, 3, 2, 1]

# Plotting
plt.plot(x, y1, label='Line 1', color='red', linewidth=2)
plt.plot(x, y2, label='Line 2', color='blue', linewidth=3)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Lines Plot')
plt.legend()
plt.show()
```

```python
#Q6

x = [11, 22, 33, 44, 55]
y = [2, 3, 1, 4, 5]
plt.grid()
plt.scatter(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot')
plt.show()

```

```python
#Q7

fig, ax = plt.subplots()
ax.scatter([1, 2, 3, 4], [13, 24, 35, 46])
plt.grid()
ax.set_title('Subplots Scatter Plot')
plt.show()
```

```python
#Q8

import numpy as np

# Generate random data
np.random.seed(0)
x = np.random.rand(10)
y = np.random.rand(10)

# Plotting
plt.grid()
plt.scatter(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Scatter Plot')
plt.show()
```

```python
#Q9

products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [20, 30, 40, 50]

# Plotting
plt.bar(products, sales,color='red',edgecolor='black')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.show()
```

```python
#Q10

# Generate random dataset
np.random.seed(0)
data = np.random.randn(100)

# Plotting
plt.hist(data, bins=3,color='yellow',edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
#Q11

expenses = ['price', 'Food', 'aplliances', 'vehicles', 'Entertainment']
amounts = [100, 60, 30, 40, 20]
mexpl=[0.01,0.01,0.01,0.01,0.01]
# Plotting
plt.pie(amounts, labels=expenses, autopct='%1.1f%%',explode=mexpl)
plt.title('Expense Distribution')
plt.show()
```
